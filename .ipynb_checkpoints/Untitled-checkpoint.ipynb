```python
from torch_geometric.datasets import QM9

import torch_geometric.transforms as T

# Example: Selecting a specific target (e.g., index 0)
class MyTransform:
    def __call__(self, data):
        data.y = data.y[:, 0]  # Select the first property
        return data

dataset = QM9(root='/tmp/QM9', transform=MyTransform())
```
